```python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from collections import deque

import tensorflow as tf
from sonnet.python.modules.base import AbstractModule
from sonnet.python.modules.basic import Linear

from environment_version_001 import env
```

```python
class Agent_DQN(object):
    def __init__(
        self,
        action_space,
        num_features= 58, 
        epsilon_greedy= 0.1, 
        gamma= 0.9,
        learning_rate= 1e-2,
        batch_size= 128, 
        memory_size= 512): 
        
        self.action_space = action_space
        self.num_actions = len(self.action_space)
        self.num_features = num_features
        self.greedy_policy = 1. - epsilon_greedy
        self.batch_size = batch_size
        self.cache = deque(maxlen= memory_size)
        
        self.step_counter = 0
        self.target_network_weight_update_iteration = 50
        
        self.state = tf.placeholder(tf.float32, shape=[None, num_features], name='state')
        self.action = tf.placeholder(tf.int32, shape=[None,], name='action')
        self.reward = tf.placeholder(tf.float32, shape=[None,], name='reward')
        self.next_state = tf.placeholder(tf.float32, shape=[None, num_features], name='next_state')
        
        self.Q_eval = self._bulid_Q_Approximate_Network('Q_eval')
        self.Q_next = self._bulid_Q_Approximate_Network('Q_next')
        self.target = self.reward + gamma * tf.reduce_max(self.Q_next, axis=1)
        state_action_index = \
        tf.stack([tf.range(tf.shape(self.state)[0], dtype=tf.int32), self.action], axis=1)
        optimal_policy_action = \
        tf.gather_nd(params=self.Q_eval, indices= state_action_index)
        
        loss = tf.reduce_mean(tf.squared_difference(self.target, optimal_policy_action))
        self.train_op = tf.train.AdamOptimizer(learning_rate).minimize(loss)
        self.sess = tf.Session()
        self.sess.run(tf.global_variables_initializer())    
        
    def _bulid_Q_Approximate_Network(self, name):
        with tf.variable_scope(str(name)):
            l1 = tf.nn.sigmoid(Linear(100, name= 'l1')(self.state))
            l2 = tf.nn.relu(Linear(100, name= 'l2')(l1))
            return Linear(self.num_actions, name='final')(l2)
        
    def action_choose(self, observation):
        state = np.expand_dims(observation, axis=0)
        if np.random.uniform() < self.greedy_policy:
            Q = self.sess.run(self.Q_eval, feed_dict={self.state:state})
            action = np.argmax(Q, axis=1)[0]
        else:
            action = np.random.randint(0, self.num_actions)
        return action
    
    def _update_cache(self, state, action, reward, next_state, done):
        self.step_counter += 1
        self.cache.append((state, action, reward, next_state, done))
    
    def learn(self):
        if self.step_counter % self.target_network_weight_update_iteration == 0:
            target_params = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='Q_next')
            eval_params = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='Q_eval')
            self.sess.run([tf.assign(t,e) for t,e in zip(target_params, eval_params)])

        minibatch = random.sample(self.cache, self.batch_size)
        state_batch = [i[0] for i in minibatch]
        action_batch = [i[1] for i in minibatch]
        reward_batch = [i[2] for i in minibatch]
        next_state_batch = [i[3] for i in minibatch]
        
        state_batch = np.array(state_batch)
        action_batch = np.squeeze(np.array(action_batch))
        reward_batch = np.array(reward_batch)
        next_state_batch = np.array(next_state_batch)

        self.sess.run(self.train_op, 
                      feed_dict={self.state:state_batch, 
                                 self.action:action_batch, 
                                 self.reward:reward_batch, 
                                 self.next_state:next_state_batch})        
        
```

```python
tf.reset_default_graph()
action_space = ['long', 'short', 'keep', 'close']
agent = Agent_DQN(action_space)
```

```python
set_values = pd.DataFrame()
set_reward = pd.DataFrame()
set_action = pd.DataFrame()

for episode in range(500):
    print (episode)
    env.reset()
    state = env.get_initial_state()
    counter = 0
    value_list = []
    reward_list = []
    action_list = []
    
    while True:
        counter +=1        
        action = agent.action_choose(state)
        reward, next_state, done = env.step(action)
        agent._update_cache(state, action, reward, next_state, done)
        
        value_list.append(env.total_value)
        reward_list.append(reward)
        action_list.append(action)
        
        
        if counter > 512:
            if counter % 100 ==0:
                agent.learn()
        if done:
            break   
    set_values[str(episode)] = value_list
    set_reward[str(episode)] = reward_list
    set_action[str(episode)] = action_list
```

```python
pd.Series(set_values.iloc[:, -1]).plot(figsize=(16,6))
```
